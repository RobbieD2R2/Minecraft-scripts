```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import timestring
import seaborn as sns
import folium


%matplotlib inline

df = pd.read_csv('C:/Users/RobbieD/Desktop/cleaned_311_test_2015.csv')

# Reduce the list of complaints to only the first 10 results
#df = df.head(25)

# Print only a small sample (10 records) for us to view below
# print(df.head(5).to_string())


print("Total number of non-null records:", len(df.index))
```

```python
# Count complaints by borough

borough_count = sns.countplot(x="borough", data=df)
```

```python
# Compare the types of complaints (and only display the top 15 types because there were dozens of them)

type_count = sns.countplot(x="complaint_type", data=df, order=df.complaint_type.value_counts().iloc[:15].index)

# Rotate the ticklabels so they don't overlap

type_count.set_xticklabels(type_count.get_xticklabels(), rotation=40, ha="right")
```

```python
# Compare which NYC agencies handled the most complaints

agency_count = sns.countplot(x="agency", data=df)

# Rotate the ticklabels so they don't overlap

agency_count.set_xticklabels(agency_count.get_xticklabels(), rotation=40, ha="right")
```
